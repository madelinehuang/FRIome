```python
#T-Test code taken from 
#https://machinelearningmastery.com/how-to-code-the-students-t-test-from-scratch-in-python/

import csv
from math import sqrt
from numpy import mean
from scipy.stats import sem
from scipy.stats import t

# function for calculating the t-test for two independent samples
def independent_ttest(data1, data2, alpha):
    # calculate means
    mean1, mean2 = mean(data1), mean(data2)
    
    # calculate standard errors
    se1, se2 = sem(data1), sem(data2)
    
    # standard error on the difference between the samples
    sed = sqrt(se1**2.0 + se2**2.0)
    
    # calculate the t statistic
    t_stat = (mean1 - mean2) / sed
    
    # degrees of freedom
    df = len(data1) + len(data2) - 2
    
    # calculate the critical value
    cv = t.ppf(1.0 - alpha, df)
    
    # calculate the p-value
    p = (1.0 - t.cdf(abs(t_stat), df)) * 2.0
    
    # return everything
    return t_stat, df, cv, p


#Put csv results into a 2d array and a phylum array
results = []
phylum = []

with open("Gender.csv") as csvfile:
    reader = csv.reader(csvfile) 
    phylum = next(reader)
    for row in reader:
        results.append(row)

phylum.remove("Male")
phylum.remove("Female")

final = [list( map(int,i) ) for i in results] 
# print(results)
# print(phylum)
# print(final)

index = 2
sig_phylum = []
bacteria_list = []
for phy in phylum:    
    print(phy)
    test_passed = 0
    
    #separate each phylum data by gender
    male = []
    female = []

    for row in final:
        if row[0] is 1: #male
            male.append(row[index])
        else: #female
            female.append(row[index])

    # calculate the t test
    alpha = 0.05
    t_stat, df, cv, p = independent_ttest(male, female, alpha)
    print('t=%.3f, df=%d, cv=%.3f, p=%.3f' % (t_stat, df, cv, p))
    
    # interpret via critical value
#     if abs(t_stat) <= cv:
#         print('Accept null hypothesis that the means are equal.')
#     else:
#         print('Reject the null hypothesis that the means are equal.')
    if abs(t_stat) > cv:
        test_passed += 1
        
    # interpret via p-value
#     if p > alpha:
#         print('Accept null hypothesis that the means are equal.')
#     else:
#         print('Reject the null hypothesis that the means are equal.')
    if p <= alpha:
        test_passed += 1
        
    print("tests passed: ", test_passed)
    
    if(test_passed is 2):
        sig_phylum.append(phy)
        bacteria_list.append(index - 2)
        
    index += 1;

print("\nphylums that are tested to be 95% confident of independent: \n", sig_phylum, 
      "\ncorresponding phylum indices: \n", bacteria_list)
```
